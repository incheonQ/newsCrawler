```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import sqlite3
import webbrowser
```

```python
def insertDB(df):
    # 데이터베이스 최신화
    con = sqlite3.connect('./db/news.db')
    cur = con.cursor()
    for i in df.itertuples():
        q = """
        INSERT into news(titles, url, type, editor, dates, publisher) VALUES (?,?,?,?,?,?)
        """
        try:
            cur.execute(q, (i[1], i[2], i[3], i[4], i[5], i[6]))
        except:
            continue
    
    con.commit()
    con.close()

    
def loadDB(pageVar, numOfcontents):
    con = sqlite3.connect('./db/news.db')
    cur = con.cursor()
    q = f"SELECT * FROM news LIMIT {pageVar * numOfcontents}, {numOfcontents}"
    tmp = pd.read_sql_query(q, con)
    con.close()
    return tmp
```

```python
def KLN():
    # KLN 물류신문

    # html page 요청 (불러오기)
    page = requests.get(f'https://www.klnews.co.kr/news/articleList.html?page={pageVar+1}&total=67684&box_idxno=&view_type=sm')
    soup = bs(page.text, "html.parser")

    # 필요한 정보 획득
    titles = soup.find_all('h4', 'titles')
    dates = soup.find_all('span', 'byline')
    
    titles_col = []
    url_col = []
    type_col = []
    editor_col = []
    dates_col = []
    publisher_col = []

    for i, j in zip(titles[:-1], dates[:-1]):
        titles_col.append(i.get_text())
        url_col.append("https://www.klnews.co.kr"+i.a['href'])          
        info = j.find_all('em')
        type_col.append(info[0].get_text())
        editor_col.append(info[1].get_text())
        dates_col.append(info[2].get_text())
        publisher_col.append('KLN')

    # 데이터프레임화
    KLN_df = pd.DataFrame()
    KLN_df['titles'] = titles_col
    KLN_df['url'] = url_col
    KLN_df['type'] = type_col
    KLN_df['editor'] = editor_col
    KLN_df['dates'] = dates_col
    KLN_df['publisher'] = publisher_col
    
    return KLN_df

def SNN():
    # 쉬핑뉴스넷

    # html page 요청 (불러오기)
    page = requests.get(f'https://www.shippingnewsnet.com/news/articleList.html?page={pageVar+1}&total=49318&box_idxno=&view_type=sm')
    soup = bs(page.text, "html.parser")

    # 필요한 정보 획득
    titles = soup.find_all('h4', 'titles')
    dates = soup.find_all('span', 'byline')
    
    titles_col = []
    url_col = []
    type_col = []
    editor_col = []
    dates_col = []
    publisher_col = []

    for i, j in zip(titles[:-1], dates[:-1]):
        titles_col.append(i.get_text())
        url_col.append("https://www.shippingnewsnet.com"+i.a['href'])          
        info = j.find_all('em')
        type_col.append(info[0].get_text())
        editor_col.append(info[1].get_text())
        dates_col.append(info[2].get_text())
        publisher_col.append('SNN')

    # 데이터프레임화
    SNN_df = pd.DataFrame()
    SNN_df['titles'] = titles_col
    SNN_df['url'] = url_col
    SNN_df['type'] = type_col
    SNN_df['editor'] = editor_col
    SNN_df['dates'] = dates_col
    SNN_df['publisher'] = publisher_col
    
    return SNN_df

def KSG():
    # 코리아쉬핑가제트 : https://www.ksg.co.kr/news/main_news.jsp
    pass

def LM():
    # 물류매거진 
    
    page = requests.get(f'http://ulogistics.co.kr/test/board.php?board=all2&no=&command=list&page={pageVar+1}')
    soup = bs(page.content.decode('euc-kr','replace'), "html.parser")

    titles_col = []
    url_col = []
    type_col = []
    editor_col = []
    dates_col = []
    publisher_col = []

    for i in soup.find_all('td', 'list_subject'):
        type_col.append(i.get_text().split('>')[0].split('<')[1].strip()) 
        url_col.append("ulogistics.co.kr/test/"+i.find('a')['href']) 
        titles_col.append(i.find('a', 'subjectColor').get_text().strip()) 
        dates_col.append(i.get_text().split(']')[-1].strip()) 
        editor_col.append('물류매거진 취재부')
        publisher_col.append('LM')

    # 데이터프레임화
    LM_df = pd.DataFrame()
    LM_df['titles'] = titles_col
    LM_df['url'] = url_col
    LM_df['type'] = type_col
    LM_df['editor'] = editor_col
    LM_df['dates'] = dates_col
    LM_df['publisher'] = publisher_col
    
    return LM_df
```

```python
def pageShow():
    global df_news
    df_news = loadDB(pageVar, numOfcontents)
    for i in range(numOfcontents):
        try:
            listbox.insert(i, f"[{df_news['publisher'][i]}] , {df_news['dates'][i]}, {df_news['titles'][i]}")
        except:
            listbox.insert(i, "기사가 없습니다!")
            
    listbox.pack()
    
def clearListbox():
    listbox.delete(0,numOfcontents-1)

def nextPage():
    global pageVar
    clearListbox()
    pageVar += 1
    pageShow()
   
def prevPage():
    global pageVar
    clearListbox()
    if pageVar > 0:
        pageVar -= 1
    pageShow()
    
    
def update():
    insertDB(KLN())
    insertDB(SNN())
#     insertDB(KSG())
    insertDB(LM())

def updateNews():
    global pageVar
    pageVar = 0
    
    update()
    clearListbox()
    pageShow()

def applyCondition():
    print(etr_keywordEntry.get()) # 키워드가 포함된 기사 추출
    
def clear(event):
    etr_keywordEntry.delete(0, len(etr_keywordEntry.get()))

def openUrl():
    if listbox.curselection() == ():
        print(None) ## 에러메시지 : 기사를 선택해주세요.
    else:
        idx = listbox.curselection()
        url = df_news.iloc[idx]['url']
        webbrowser.open(url)
```
